```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
df.head() 
```

```python
df.info()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set()
fig , axes = plt.subplots(1,2,figsize=(20,10))
sns.barplot(data=df.groupby('Sub-Category')['Sales','Profit'].agg(sum) , x = 'Sales' , 
            y = 'Profit' , ax = axes[1])
df.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind = 'bar' , ax = axes[0])
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))
sns.barplot(df['Segment'],df['Sales'],ax=axes[0])
sns.barplot(df['Ship Mode'],df['Sales'],ax=axes[1])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,5))
df.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind = 'bar' , ax = axes[0]).set_title('RELATION BETWEEN DISCOUNT AND PROFIT')
df.groupby('Sub-Category')['Sales','Quantity'].agg(sum).plot(kind = 'bar' , ax = axes[1]).set_title('RELATION BETWEEN SALES AND QUANTITY')
plt.xticks(rotation = 90)
plt.show()
```

```python
fig.ax = plt.subplots(1,1 , figsize = (12,6))
sns.countplot(df['Quantity'] , hue = df['Region'])
plt.show()
```

```python

fig,axes=plt.subplots(2,2,figsize = (20,10))
sns.distplot(df['Sales'],ax=axes[0,0])
sns.distplot(df['Quantity'],ax=axes[0,1])
sns.distplot(df['Profit'],ax=axes[1,0])
sns.distplot(df['Discount'],ax=axes[1,1])
plt.show()
```
